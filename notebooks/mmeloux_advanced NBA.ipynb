```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
players = pd.read_csv("../nba_forecast/data/players.csv") 
```

```python
players
```

```python
players_dict = players.to_dict(orient='records')
```

```python
third_season_advanced = []

for player in players_dict:
    response = requests.get(player['url'])
    soup = BeautifulSoup(response.content, "html.parser")
    
    try:
        advanced_season3 = soup.find('div',id="div_advanced").find_all('tr',class_="full_table")[2]


        season = advanced_season3.find('th').find('a').string

        for td in advanced_season3.find_all('td'):
            if td['data-stat'] == "age":
                age = td.string
            elif td['data-stat'] == "team_id":
                team_id = td.string
            elif td['data-stat'] == "pos":
                pos = td.string
            elif td['data-stat'] == "g":
                games = td.string
            elif td['data-stat'] == "mp":
                minutes_played = td.string
            elif td['data-stat'] == "per":
                per = td.string
            elif td['data-stat'] == "ts_pct":
                ts_pct = td.string
            elif td['data-stat'] == "fg3a_per_fga_pct":
                fg3a_per_fga_pct = td.string
            elif td['data-stat'] == "fta_per_fga_pct":
                fta_per_fga_pct = td.string
            elif td['data-stat'] == "orb_pct":
                orb_pct = td.string
            elif td['data-stat'] == "drb_pct":
                drb_pct = td.string
            elif td['data-stat'] == "trb_pct":
                trb_pct = td.string
            elif td['data-stat'] == "ast_pct":
                ast_pct = td.string
            elif td['data-stat'] == "stl_pct":
                stl_pct = td.string
            elif td['data-stat'] == "blk_pct":
                blk_pct = td.string
            elif td['data-stat'] == "tov_pct":
                tov_pct = td.string
            elif td['data-stat'] == "usg_pct":
                usg_pct = td.string
            elif td['data-stat'] == "ows":
                ows = td.string
            elif td['data-stat'] == "dws":
                dws = td.string
            elif td['data-stat'] == "ws":
                ws = td.string
            elif td['data-stat'] == "ws_per_48": 
                ws_per_48 = td.string
            elif td['data-stat'] == "obpm":
                obpm = td.string
            elif td['data-stat'] == "dbpm":
                dbpm = td.string
            elif td['data-stat'] == "bpm":
                bpm = td.string
            elif td['data-stat'] == "vorp":
                vorp = td.string

        third_season_advanced.append(
            {
            'name':player['player'],
            'season':season,
            'age':age,
            'team_id':team_id,
            'pos':pos,
            'games':games,
            'minutes_played':minutes_played,
            'per':per,
            'ts_pct':ts_pct,
            'fg3a_per_fga_pct':fg3a_per_fga_pct,
            'fta_per_fga_pct':fta_per_fga_pct,
            'orb_pct':orb_pct,
            'drb_pct':drb_pct,
            'trb_pct':trb_pct,
            'ast_pct':ast_pct,
            'stl_pct':stl_pct,
            'blk_pct':blk_pct,
            'tov_pct':tov_pct,
            'usg_pct':usg_pct,
            'ows':ows,
            'dws':dws,
            'ws':ws,
            'ws_per_48':ws_per_48,
            'obpm':obpm,
            'dbpm':dbpm,
            'bpm':bpm,
            'vorp':vorp,
            'uni_url':player['uni_url']}
        )
        
    except (IndexError, AttributeError):
        continue
```

```python
third_season_advanced_df = pd.DataFrame(third_season_advanced)
third_season_advanced_df
```

```python
third_season_advanced_df.to_csv('../nba_forecast/data/third_season_advanced.csv', index=False)
```
