```python
import pandas as pd
import numpy as np
```

```python
uni_data_df = pd.read_csv('../nba_forecast/data/df_donne_universitaire.csv')
# uni_data_df = uni_data_df.iloc[: , 1:] #drop first column as csv contains indexes already
#drop dummy columns (empty)
uni_data_df = uni_data_df.drop(columns=['ws-dum', 'bpm-dum'])
```

```python
uni_data_df = uni_data_df.drop_duplicates()
```

```python
uni_data_df.head()
```

```python
years_df = uni_data_df.groupby(['player_name','school_name']).size().to_frame('years').reset_index()
```

```python
final_df = pd.merge(uni_data_df,years_df,how='inner',on=['player_name','school_name'])
```

```python
final_df = final_df.drop_duplicates(['player_name'], keep='last').reset_index().drop(columns='index')
```

```python
player_id_list = []

for url in final_df['uni_url'].tolist():
    id = url.split('/')[-1][:-5]
    player_id_list.append(id)
    
final_df['player_id'] = pd.Series(player_id_list, index=final_df.index)
final_df=final_df.iloc[: , 1:]
```

```python
final_df.to_csv('../nba_forecast/data/last_ncaa_season.csv', index=False)
```
