## Scrapping conference ratings

```python
conf_ratings = []

conf_rating_urls = [
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2012-04-30',
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2013-04-30',
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2014-04-30',
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2015-04-30',
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2016-04-30',
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2017-04-30',
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2018-04-30',
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2019-04-30',
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2020-04-30',
    'https://www.teamrankings.com/ncaa-basketball/rpi-ranking/rpi-rating-by-conf?date=2021-04-30'
]

for url in conf_rating_urls:
    conf_response = requests.get(url)
    conf_soup = BeautifulSoup(conf_response.content, "html.parser")
#     conf_comments=conf_response.find_all(string=lambda text:isinstance(text,Comment))
    conf_table = conf_soup.find(class_='tr-table datatable').find('tbody').find_all('tr')
    
    for tr in conf_table:
        season = f'{int(url[-10:-6])-1}-{url[-8:-6]}'
        conf = tr.find('td',class_='nowrap').string
        if conf == 'Atlantic 10':
            conf='A-10'
        elif conf == 'Ohio Valley':
            conf = 'OVC'
        elif conf == 'Mountain West':
            conf = 'MWC'
        elif conf == 'Horizon League':
            conf = 'Horizon'
        elif conf == 'MAAC':
            conf = 'AAC'
        rating = float(tr.find('td',class_='text-right').get('rel'))

        conf_ratings.append(
        {
        'season':season,
        'conference':conf,
        'rating':rating,
    }
    )
```

```python
conf_ratings_df = pd.DataFrame(conf_ratings)
conf_ratings_df.to_csv('../nba_forecast/data/conf_ratings.csv', index=False)
```

```python
team_urls = [
    'https://www.basketball-reference.com/teams/BOS/2021.html',
    'https://www.basketball-reference.com/teams/BRK/2021.html',
    'https://www.basketball-reference.com/teams/TOR/2021.html',
    'https://www.basketball-reference.com/teams/PHI/2021.html',
    'https://www.basketball-reference.com/teams/NYK/2021.html',
    'https://www.basketball-reference.com/teams/CHI/2021.html',
    'https://www.basketball-reference.com/teams/CLE/2021.html',
    'https://www.basketball-reference.com/teams/DET/2021.html',
    'https://www.basketball-reference.com/teams/IND/2021.html',
    'https://www.basketball-reference.com/teams/MIL/2021.html',
    'https://www.basketball-reference.com/teams/WAS/2021.html',
    'https://www.basketball-reference.com/teams/ATL/2021.html',
    'https://www.basketball-reference.com/teams/ORL/2021.html',
    'https://www.basketball-reference.com/teams/MIA/2021.html',
    'https://www.basketball-reference.com/teams/CHO/2021.html',
    'https://www.basketball-reference.com/teams/MIN/2021.html',
    'https://www.basketball-reference.com/teams/OKC/2021.html',
    'https://www.basketball-reference.com/teams/DEN/2021.html',
    'https://www.basketball-reference.com/teams/POR/2021.html',
    'https://www.basketball-reference.com/teams/UTA/2021.html',
    'https://www.basketball-reference.com/teams/LAL/2021.html',
    'https://www.basketball-reference.com/teams/LAC/2021.html',
    'https://www.basketball-reference.com/teams/GSW/2021.html',
    'https://www.basketball-reference.com/teams/PHO/2021.html',
    'https://www.basketball-reference.com/teams/SAC/2021.html',
    'https://www.basketball-reference.com/teams/MEM/2021.html',
    'https://www.basketball-reference.com/teams/NOP/2021.html',
    'https://www.basketball-reference.com/teams/HOU/2021.html',
    'https://www.basketball-reference.com/teams/SAS/2021.html',
    'https://www.basketball-reference.com/teams/DAL/2021.html'
]

team_rankings = []

for url in team_urls:

    team_response = requests.get(url)
    team_soup = BeautifulSoup(team_response.content, "html.parser")
    team_comments=team_soup.find_all(string=lambda text:isinstance(text,Comment))
    
    for p in team_soup.find('div',id='info',class_='teams').find_all('p'):
        if p.find('a') and p.find('a').get('href') == "/about/glossary.html#wins_pyth":
            ratio = p.get_text().split(':')[1].strip()[:5]
            wins = int(ratio[:2])
            games = int(ratio[:2]) + int(ratio[3:5])
            win_rate = wins/games
            break
    
    for comment in team_comments:
        comment=BeautifulSoup(str(comment), 'html.parser')
        table = comment.find('div',class_='table_container', id="div_team_misc")
        if table:
            for tr in table.find('tbody').find_all('tr'):
                if tr.find('th').string == 'Team':
                    for td in tr.find_all('td'):
                        if td.get('data-stat') == 'off_rtg':
                            off_rtg = td.string
                        elif td.get('data-stat') == 'def_rtg':
                            def_rtg = td.string
                        elif td.get('data-stat') == 'attendance':
                            attendance = td.string
                        elif td.get('data-stat') == 'pace':
                            pace = td.string
                if tr.find('th').string == 'Lg Rank':
                    for td in tr.find_all('td'):
                        if td.get('data-stat') == 'off_rtg':
                            off_rank = td.string
                        elif td.get('data-stat') == 'def_rtg':
                            def_rank = td.string

            team_rankings.append(
        {
        'Team':url[-13:-10],
        'off_rtg':off_rtg,
        'def_rtg':def_rtg,
        'off_rank':off_rank,
        'def_rank':def_rank,
        'pace':pace,
        'attendance':attendance,
        'expected_winrate':round(float(win_rate),2)
        })
```

```python
team_rankings_df = pd.DataFrame(team_rankings)
team_rankings_df.to_csv('../nba_forecast/data/team_stats_2021.csv', index=False)
```

```python
univ_data = []

for player in players_dict:

    if not str(player['uni_url']).startswith('http'):
        continue
    player_id = ' '
    if str(player['uni_url']).startswith('http'):
        uni_url = player['uni_url']
        player_id = uni_url.split('/')[-1][:-5]

    response = requests.get(player['uni_url'])
    soup = BeautifulSoup(response.content, "html.parser")
    
    comments=soup.find_all(string=lambda text:isinstance(text,Comment))
    
    for comment in comments:
        advanced = False
        comment=BeautifulSoup(str(comment), 'html.parser')
        advanced = comment.find('div', id='div_players_advanced')
        if advanced:
            last_year=advanced.find_all('tr')[-1]


    try:

        season = '2020-2021'

        for td in last_year.find_all('td'):
            if td['data-stat'] == "school_name":
                school_name = td.string
            elif td['data-stat'] == "pos":
                pos = td.string
            elif td['data-stat'] == "g":
                games = td.string
            elif td['data-stat'] == "gs":
                games_started = td.string
            elif td['data-stat'] == "mp":
                minutes_played = td.string
            elif td['data-stat'] == "per":
                per = td.string
            elif td['data-stat'] == "ts_pct":
                ts_pct = td.string
            elif td['data-stat'] == "fg3a_per_fga_pct":
                fg3a_per_fga_pct = td.string
            elif td['data-stat'] == "fta_per_fga_pct":
                fta_per_fga_pct = td.string
            elif td['data-stat'] == "orb_pct":
                orb_pct = td.string
            elif td['data-stat'] == "drb_pct":
                drb_pct = td.string
            elif td['data-stat'] == "trb_pct":
                trb_pct = td.string
            elif td['data-stat'] == "ast_pct":
                ast_pct = td.string
            elif td['data-stat'] == "stl_pct":
                stl_pct = td.string
            elif td['data-stat'] == "blk_pct":
                blk_pct = td.string
            elif td['data-stat'] == "tov_pct":
                tov_pct = td.string
            elif td['data-stat'] == "usg_pct":
                usg_pct = td.string
            elif td['data-stat'] == "ows":
                ows = td.string
            elif td['data-stat'] == "dws":
                dws = td.string
            elif td['data-stat'] == "ws":
                ws = td.string
            elif td['data-stat'] == "ws_per_48": 
                ws_per_48 = td.string
            elif td['data-stat'] == "ws_per_40": 
                ws_per_40 = td.string
            elif td['data-stat'] == "obpm":
                obpm = td.string
            elif td['data-stat'] == "dbpm":
                dbpm = td.string
            elif td['data-stat'] == "bpm":
                bpm = td.string
            elif td['data-stat'] == "vorp":
                vorp = td.string

        univ_data.append(
            {
            'player_id':player_id,
            'name':player['player'],
            'season':season,
            'school_name':school_name,
            'g':games,
            'gs':games_started,
            'mp':minutes_played,
            'per':per,
            'ts_pct':ts_pct,
            'fg3a_per_fga_pct':fg3a_per_fga_pct,
            'fta_per_fga_pct':fta_per_fga_pct,
            'orb_pct':orb_pct,
            'drb_pct':drb_pct,
            'trb_pct':trb_pct,
            'ast_pct':ast_pct,
            'stl_pct':stl_pct,
            'blk_pct':blk_pct,
            'tov_pct':tov_pct,
            'usg_pct':usg_pct,
            'ows':ows,
            'dws':dws,
            'ws':ws,
            'obpm':obpm,
            'dbpm':dbpm,
            'bpm':bpm,
            'uni_url':player['uni_url']}
        )
        
    except (IndexError, AttributeError):
        continue
```
