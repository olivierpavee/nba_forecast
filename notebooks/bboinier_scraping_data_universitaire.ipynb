```python
import pandas as pd
import time
```

```python
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
```

```python
df_players = pd.read_csv('../nba_forecast/data/players.csv')
```

```python
list_url = list(df_players['uni_url'].values)
```

```python
driver_options = webdriver.ChromeOptions()

prefs = {"download.default_directory" : "/mnt/c/Users/boris/Desktop/Brouillon/Entrainement_Selenium"}
driver_options.add_experimental_option("prefs",prefs)
driver = webdriver.Chrome(chrome_options=driver_options)
driver.maximize_window()
```

```python
pandas_df = pd.DataFrame()
agree = False
dico_stats = {'uni_url': [],'player_name': [], 'season': []}

for url in list_url:

    time.sleep(2)
    
    try:
        driver.get(url)
    except:
        print(f'{url} invalide')
          
    time.sleep(2)

    if agree == False:
        agree_button = driver.find_element_by_xpath('//*[@id="qc-cmp2-ui"]/div[2]/div/button[3]')
        agree_button.click()
        agree = True

    player_name = driver.find_element_by_xpath(
        '//*[@id="meta"]/div/h1/span').get_attribute("innerHTML").splitlines()[0]

    raw_stats = driver.find_element_by_xpath(
        "//*[@id='players_advanced']/tbody").get_attribute(
            "innerHTML").splitlines()

    for raw_stat in raw_stats:
        soup_string = BeautifulSoup(raw_stat)
        # print(soup_string.prettify())
        # print(soup_string.find_all('td'))
        dico_stats['player_name'].append(player_name)
        dico_stats['season'].append(soup_string.find('th').text)
        print(soup_string.find('th').text)
        for ele in soup_string.find_all('td'):
            # print(ele['data-stat'])
            # print(ele.text)
            if ele['data-stat'] not in dico_stats.keys():
                dico_stats[ele['data-stat']] = []
                dico_stats[ele['data-stat']].append(ele.text)
            else:
                dico_stats[ele['data-stat']].append(ele.text)

driver.quit()
```

```python
dico_stats.pop('uni_url', None)
```

```python
pandas_df = pd.DataFrame.from_dict(dico_stats)
```
