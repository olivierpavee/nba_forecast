```python
driver = webdriver.Chrome()
driver.maximize_window()

agree = False
consti_dico = False
dico_scraping = {'year':[]}
key_index_list = []


list_url = [f'https://www.nba.com/stats/draft/combine-anthro/?SeasonYear={year}' for year in [f'201{x}-1{x+1}' for x in range(2,9)]]


for url in list_url:
    
    year = url[-7:]

    time.sleep(2)
    
    try:
        driver.get(url)
    except:
        print(f'{url} invalide')
          
    time.sleep(2)

    if agree == False:
        agree_button = driver.find_element_by_xpath('//*[@id="onetrust-accept-btn-handler"]')
        agree_button.click()
        agree = True
        
    tableau = ' '.join(driver.find_element_by_xpath(
         '/html/body/main/div/div/div[2]/div/div/nba-stat-table/div[2]/div[1]/table').get_attribute("innerHTML").splitlines())
    

    soup_tableau = BeautifulSoup(tableau)

    
    if consti_dico == False:
        for elem in soup_tableau.find_all('th'):
            print(elem['data-field'])
            dico_scraping[elem['data-field']] = []
        consti_dico = True
        key_index_list = list(dico_scraping)
    
        
    for elem in soup_tableau.find_all('tr')[1:]:
        i=1
        dico_scraping['year'].append(year)
        
        for elem_2 in elem.find_all('td'):
            print(elem_2.text)
            dico_scraping[key_index_list[i]].append(elem_2.text)
            i += 1
        

#     player_name = driver.find_element_by_xpath(
#         '//*[@id="meta"]/div/h1/span').get_attribute("innerHTML").splitlines()[0]

#     raw_stats = driver.find_element_by_xpath(
#         "//*[@id='players_advanced']/tbody").get_attribute(
#             "innerHTML").splitlines()

#     for raw_stat in raw_stats:
#         soup_string = BeautifulSoup(raw_stat)
#         # print(soup_string.prettify())
#         # print(soup_string.find_all('td'))
#         dico_stats['player_name'].append(player_name)
#         dico_stats['season'].append(soup_string.find('th').text)
#         print(soup_string.find('th').text)
#         for ele in soup_string.find_all('td'):
#             # print(ele['data-stat'])
#             # print(ele.text)
#             if ele['data-stat'] not in dico_stats.keys():
#                 dico_stats[ele['data-stat']] = []
#                 dico_stats[ele['data-stat']].append(ele.text)
#             else:
#                 dico_stats[ele['data-stat']].append(ele.text)

driver.quit()
```
